```python
import pandas as pd
import random
from faker import Faker
from random import randrange
from datetime import datetime

nr_of_businessuser = 200

fake = Faker('de_DE')

businessuser = []

for businessuser_id in range(nr_of_businessuser):

    #create Business User's name
    name = fake.first_name()
    name2 = fake.last_name()

    # Create email 
    email = fake.ascii_email()

    # Create How would like to submit your project brief?
    projectbrief = random.choice(["I'll do it myself", "Book a call with MARQETR"])

    # How soon do you need a marketing experts?
    need = random.choice(["Immediately", "In 1 -3 weeks", "In 3+ weeks", "Not sure yet. Decide later"])

    # Do you have any location preference for your potential marketing experts?
    locationPreference = random.choice(["I prefer marketing experts in my city only", "I prefer marketing experts in my country only", 
                                        "I prefer marketing experts within the same time zone of me",
                                        "I prefer marketing experts within 3 time zones of me",
                                        "I don't care where my marketing experts are located"
                                        ])
    
    # How long do you need the marketing expert for?
    howLong = random.choice(["<1 month", 
                             "1 - 3 months", 
                              "3 - 6 months",
                              "6 - 9 months",
                              "9+ months",
                             'Not sure yet. Decide later'])
    
    # What is the preferred language that you'd like to communicate in with your marketing expert?
    language = random.choice(["English", 
                             "Bahasa", 
                              "Mandarin"])
    
    # Specify probabilities of each category (must sum to 1.0)
    weights = [0.6, 0.2, 0.1]
    
    # Which of the following most accurately describe(s) you?
    gender = random.choice(["Female", "Male", "Transgender Female", "Transgender Male", "Non-Binary" , "Intersex", "I Prefer Not to Say"])

    businessuser.append([name, name2, email,projectbrief, need, locationPreference, howLong, (language,weights), gender])


businessuser_df = pd.DataFrame(businessuser, columns=['Your first name', 
                                                      'Your last name', 
                                                      'Your registered email address on MARQETR is...',
                                                      'How would like to submit your project brief?',
                                                      'How soon do you need a marketing experts?',
                                                      'Do you have any location preference for your potential marketing experts?',
                                                      'How long do you need the marketing expert for?',
                                                      'What is the preferred language that you do like to communicate in with your marketing expert?',
                                                      'Which of the following most accurately describe(s) you?'
                                                      ]) 
                
pd.pandas.set_option('display.max_columns', None)
print(businessuser_df)
```
